```python
#Task 1
from math import *

'''Create a new class called Point. This class will have a “x” and “y” attribute.'''

class Point():
    def __init__(self,received_x,received_y,):
        self.x=received_x
        self.y=received_y
        
    def __str__(self):
        return f"{self.x},{self.y}"
    
    def __repr__(self):
        return self.__str__()
    
'''write a function called distance_between_points that takes two Points as arguments 
and returns the distance between them.'''        

def distance_between_points(point1,point2):
    x=point1.x-point2.x
    y=point1.y-point2.y
    dis=sqrt(x**2+y**2)
    return dis

p1=Point(3,5)
p2=Point(4,6)
distance_between_points(p1,p2)
```

```python
#Task 2
'''Create a new class called Rectangle, this class will have width, height and corner attributes. The corner
attribute is an instance of the Point class created in task1.'''

class Rectangle():
    def __init__(self,wid,hei,cor):
        if not isinstance(cor,Point):
            raise ValueError("corner should be a Point instance")
        self.width=wid
        self.height=hei
        self.corner=cor
    
    def __str__(self):
        return f"{self.width},{self.height} at point({self.corner})"
    
    def __repr__(self):
        return self.__str__()
        

p=Point(0,0)
x=Rectangle(100,200,p)   
print(x)

```

```python
#Task 2 A:
'''Write a function called “find_center” that takes a Rectangle has an argument and returns a Point
that contains the coordinates of the center of the Rectangle. (Assuming the corner of the rectangle
is on the origin)'''

def find_center(rec):
    if not isinstance(rec,Rectangle):
        raise ValueError("Argument should be a rectangle instance")
    return Point(0.5*rec.width,0.5*rec.height)

p=Point(0,0)
rec=Rectangle(100,200,p)  
find_center(rec)    
    
```

```python
#Task 2 B:
'''Write a function named move_rectangle that takes a Rectangle and two numbers
named dx and dy. It should change the location of the rectangle by adding dx to 
the x coordinateof corner and adding dy to the y coordinate of corner.'''

def move_rectangle(rect,dx,dy):
    if not isinstance(rect,Rectangle):
        raise ValueError("Argument should be a rectangle instance")
    m_x=rect.corner.x+dx
    m_y=rect.corner.y+dy
    return Point(m_x,m_y)

rect=Rectangle(100,200,Point(2,5))  
move_rectangle(rect,3,4)
```

```python
#Task 4:
'''Create a Time class that has instant attributes: hour, minutes and seconds. '''

'''A. Write a __str__ method for the Time class to print an output in the format: 02:34:10
Hint: Try to use the “format” method of String '''

'''B.Write a boolean function called is_after that takes two Time objects, t1 and t2, and returns True
if t1 follows t2 chronologically and False otherwise. Challenge: don’t use an if statement'''

class Time():
    def __init__(self,hour,minutes,seconds):
        self.hour=hour
        self.minutes=minutes
        self.seconds=seconds
    
    def __str__(self):
        return f"{self.hour}:{self.minutes}:{self.seconds}"
            
def is_after(t1,t2):
    t1_sec=int(t1.hour)*3600+int(t1.minutes)*60+int(t1.seconds)
    t2_sec=int(t2.hour)*3600+int(t2.minutes)*60+int(t2.seconds)
    return t1_sec<t2_sec
        
           

print(Time('02',34,10))
t1=Time('02',34,'09')
t2=Time('02',34,'10')
print(is_after(t1,t2))

```

```python
#Q1：
'''Write an add method for Points that works with either a Point object or a tuple:
• If the second operand is a Point, the method should return a new Point whose x coordinate is the
sum of the x coordinates of the operands, and likewise for the y coordinates.
• If the second operand is a tuple, the method should add the first element of the tuple to
the x coordinate and the second element to the y coordinate, and return a new Point with the
result.'''

class Point():
    def __init__(self,received_x,received_y,):
        self.x=received_x
        self.y=received_y
        
    def __str__(self):
        return f"{self.x},{self.y}"
    
    def __repr__(self):
        return self.__str__()
    
    def __add__(self,obj):
        if isinstance(obj,Point):
            return (self.x+obj.x,self.y+obj.y)
        if isinstance(obj,tuple):
            return (self.x+obj[0],self.y+obj[1])
    


p1=Point(1,0)
p2=Point(2,3)
t1=(4,5)
print(p1.__add__(p2))
print(p1.__add__(t1))
```
